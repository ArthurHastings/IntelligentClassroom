```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q1 + 1.5 * IQR

    return df[(df[column] >= lower_lim) & (df[column] <= upper_lim)]
```

```python
num_columns = ["age", "engagement_level", "accuracy_of_responses", "emotion_score", "frustration_level", "learning_progress"]
```

```python
for col in num_columns:
    df = remove_outliers_iqr(df, col)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def label_encoder(df, columns):
    encoders = {}

    for col in columns:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
        encoders[col] = encoder

    return df, encoders
```

```python
df, encoders = label_encoder(df, columns=["task_type", "task_difficulty", "skill_gap", "progress_category"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
X = df[["task_type", "task_difficulty", "age", "engagement_level", "accuracy_of_responses", "emotion_score", "time_spent_on_task", "frustration_level", "skill_gap", "learning_progress"]]
y = df["progress_category"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
log_model = LogisticRegression()
```

```python
### LOG MODEL 1
```

```python
log_model.fit(X_train, y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix, classification_report
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
